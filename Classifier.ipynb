```python
class MyGUI:
    def __init__(self, master):
        self.valuesdata1 = StringVar()
        self.valuesdata2 = StringVar()
        self.valuesdata5 = StringVar()
        self.valuesdata6 = StringVar()
        self.valuesdata7 = StringVar()
        #########
        self.var3 = StringVar()
        self.var4 = StringVar()
        self.e1=StringVar()
        self.e2=StringVar()
        self.e3=StringVar()
        #########3
        self.value=0
        self.head=StringVar()
        self.data=[]
        self.layer=[]
        self.flag=0
        activation = ["Select",
                   "relu",
                   "sigmoid",
                   "tanh"
                ]
        self.testing=0;
        self.variable=StringVar(master)
        self.variable.set(activation[0])
        pooling = ["Select",
                   "MaxPooling",
                   "AveragePooling"
                ]
        self.layercount=0
        self.datacount=0
        self.variable1=StringVar(master)
        self.variable1.set(pooling[0])
        self.yaxis=IntVar()
        self.row=IntVar()
        self.col=IntVar()
        self.band=IntVar()
        self.epochs=IntVar()
        self.cls=IntVar()
        self.threshold=float()
        ############
        self.image=StringVar()
        self.imgread=StringVar()
        self.head=StringVar()
        self.file = open('load.txt','w')         
       ##############
        self.yaxis=100
        self.master = master
        self.master.title(" Image Classification tool ")
        self.master.config(bg="#dbdbdb")
        self.master.geometry("800x500")
        self.can = Canvas(master, width=785, height=85, bg='#20232A')        
        self.can.place(x=5, y=5)
        self.l1 = ttk.Label(master, text="Recurrent Neural Network Algorithm",font=('times new roman', 18, 'bold'),style='green/black.TLabel',background='#20232A',foreground="#dbdbdb")
        self.l1.place(x=180, y=30)
        self.can = Canvas(master, width=785, height=230, bg='#20232A')        
        self.can.place(x=5, y=95)
             
        self.l2 = ttk.Label(master,text="Define the parameters for RNN classifier :",font=('times new roman', 12, 'bold'),style='gb.TLabel',background='#20232A',foreground="#dbdbdb")
        self.l2.place(x=25, y=115, width=350)
        
        self.valuesdata1.set("Enter the Epochs value")
        self.valuesdata2.set("Enter Number Of Training Classes")
        self.valuesdata5.set("Enter the Threshold value")
        self.valuesdata6.set("Select Activation Function")
        self.valuesdata7.set("Select MaxPooling/AveragePooling Layer")
        self.l3 = ttk.Label(master,textvariable=self.valuesdata1,style='gb.TLabel',font=('times new roman', 11, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.l3.place(x=35, y=165, width=250) 
        Entry(self.master,textvariable=self.e1).place(x=485, y=165, width=80)
        self.l4 = ttk.Label(master,textvariable=self.valuesdata2,style='gb.TLabel',font=('times new roman', 11, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.l4.place(x=35, y=195, width=250) 
        Entry(self.master,textvariable=self.e2).place(x=485, y=195, width=80)
        self.ll4 = ttk.Label(master,textvariable=self.valuesdata5,style='gb.TLabel',font=('times new roman', 11, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.ll4.place(x=35, y=225, width=250) 
        Entry(self.master,textvariable=self.e3).place(x=485, y=225, width=80)
        self.ll6 = ttk.Label(master,textvariable=self.valuesdata6,style='gb.TLabel',font=('times new roman', 11, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.ll6.place(x=35, y=255, width=250) 
        self.ee3=OptionMenu(master,self.variable,*activation)
        self.ee3.place(x=485,y=255,width=80)
        self.ll6 = ttk.Label(master,textvariable=self.valuesdata7,style='gb.TLabel',font=('times new roman', 11, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.ll6.place(x=35, y=285, width=290)
        self.el3=OptionMenu(master,self.variable1,*pooling)
        self.el3.place(x=485,y=285,width=80)              
        
        self.can = Canvas(master, width=785, height=160, bg='#20232A')        
        self.can.place(x=5, y=330) 
        self.b1 = ttk.Button( text='Select ImageFile', style='green/black.TButton' , command=self.img )
        self.b1.place(x=35, y=355, width=110)
        ttk.Label(master,textvariable=self.var4,relief="sunken").place(x=35, y=395, width=300)
        self.b2 = ttk.Button( text='Select HeaderFile', style='green/black.TButton' , command=self.header)
        self.b2.place(x=395, y=355, width=110)
        ttk.Label(textvariable=self.var3,relief="sunken" ).place(x=395, y=395, width=300)        
        self.bx1 = ttk.Button( text='Next', style='green/black.TButton' , command=self.nextw)
        self.bx1.place(x=485, y=455, width=150)
        
   
        
    def header(self):
        filename =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("Header file","*.hdr"),("all files","*.*")))
        self.var3.set(filename)
        self.head=filename
        print(self.head) 
    
    def img(self):
        
        filename =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("all files","*.*"),("Text file","*.txt")))
        self.var4.set(filename)
        self.image=filename
        self.imgread=self.var4.get()
        print(self.imgread)
        
    def headerreader(self,headpath):
        i=1
        r=-1
        c=-1
        b=-1
        pat1 = re.compile(r"\brows\b", re.IGNORECASE) # but matches because pattern ignores case
        pat2 = re.compile(r"\bcols\b", re.IGNORECASE) # but matches because pattern ignores case
        pat3 = re.compile(r"\bbands\b", re.IGNORECASE) # but matches because pattern ignores case
        pat4 = re.compile(r"\bdatatype\b", re.IGNORECASE) # but matches because pattern ignores case
        lines = [] #Declare an empty list named "lines"
        with open (headpath, 'rt') as in_file: # Open file txt for reading of text data.
            for line in in_file:
                if pat1.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    r=line.split(" ",1)[1] 
                    print("this is the number of rows extracted:")
                    print(r)

                elif pat2.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    c=line.split(" ",1)[1] 
                    print("this is the number of columns extracted:")
                    print(c)

                elif pat3.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    b=line.split(" ",1)[1] 
                    print("this is the number of bands extracted:")
                    print(b)
                elif pat4.search(line) != None:
                    lines.append(line) 
                    i=i+1
                    d=line.split(" ",1)[1] 
                    d=d.split(" ",1)[0] 
                    d=d.split("U",1)[1] 
                    print("this is the datatype extracted:")
                    print(d)
                else:
                    i=i+1

        if( r==-1 or b==-1 or c==-1 or not(d)):
            print("ERROR: header information insuffficient")

        in_file.close()
        print(lines)
        rr=int(r)
        cc=int(c)
        bb=int(b)
        dd=int(d)
        return rr,cc,bb,dd
    
    
   
        
     
    def add(self):
        if len(self.data)>=int(self.e2.get()):
            messagebox.showerror("Error", "Number of classes exceeded")
        else:
            add =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("all files","*.*"),("Text file","*.txt")))
            r=self.e2.get()
            self.cls=int(r)+1
            self.data.append(add)
            self.T.config(state="normal")
            self.T.delete('1.0', END)
            for ent in self.data:
                self.T.insert(END, ent+'\n')
            self.T.config(state="disabled")
            if len(self.data)>=(self.cls-1):
                self.b3.config(state="disabled")
                self.T.config(state="disabled")
            elif len(self.data)<(self.cls-1):
                self.b3.config(state="normal")
                self.T.config(state="normal")
            self.datacount=self.datacount+1
            print(len(self.data))
            print(self.cls)
     
        
    def remove(self):
        if len(self.data)==0:
             messagebox.showerror("Error", "No data added")
        else:
            self.data=self.data[:-1]
            print(self.data)
            self.T.config(state="normal")
            self.T.delete('1.0', END)
            for ent in self.data:
                self.T.insert(END, ent+'\n')
            self.T.config(state="disabled")
            if len(self.data)!=0:
                self.b4.config(state="normal")
            else:
                self.b4.config(state="disabled")
            
            self.datacount=self.datacount-1
        
    def savedone(self):
        self.master.destroy()
    
    
    
    def addlayers(self):
       # if len(self.data) == 0:
        #    messagebox.showerror("Error", "Insert the training data")
         #else:
        self.window9.destroy()
        self.window7=ttk.tkinter.Toplevel()
        """whichNN="addLayer"
        self.window7.protocol("WM_DELETE_WINDOW",self.delete)"""
        self.window7.geometry("596x285")
        self.can = Canvas(self.window7, width=385, height=50, bg='#20232A')        
        self.can.place(x=5, y=5)
        self.l11=Label(self.window7,text="Add Layers sequentially in the RNN",font=('times new roman', 14, 'bold'),style='green/black.TLabel',background='#20232A',foreground="#dbdbdb")
        self.l11.place(x=50,y=20)
        self.can = Canvas(self.window7, width=385, height=161, bg='#20232A')        
        self.can.place(x=5, y=58.5)
        self.l22=Label(self.window7,text="Convolution Layer",style='gb.TLabel',font=('times new roman', 12, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.l22.place(x=25,y=90,width=150)
        pool=str(self.variable1.get())
        poolc="MaxPooling"
        if(pool==poolc):
            self.value=1
        else:
             self.value=2
        if(self.value is 1):
            self.l33=Label(self.window7,text="MaxPooling Layer",style='gb.TLabel',font=('times new roman', 12, 'bold'),background='#20232A',foreground="#dbdbdb")
            self.l33.place(x=25,y=125,width=150)
        else:
            self.l33=Label(self.window7,text="AveragePooling Layer",style='gb.TLabel',font=('times new roman', 12, 'bold'),background='#20232A',foreground="#dbdbdb")
            self.l33.place(x=25,y=125,width=150)
        self.l34=Label(self.window7,text="LSTM Layer",style='gb.TLabel',font=('times new roman', 12, 'bold'),background='#20232A',foreground="#dbdbdb")
        self.l34.place(x=25,y=160,width=150)
        self.b22 = ttk.Button(self.window7, text='Add', style='add.TButton' , command=self.addConv)
        self.b22.place(x=260, y=90, width=50)
        self.b44 = ttk.Button(self.window7, text='Add',  style='add.TButton' , command=self.addPool)
        self.b44.place(x=260, y=125, width=50)
        self.b55 = ttk.Button(self.window7, text='Add', style='add.TButton' , command=self.addLSTM)
        self.b55.place(x=260, y=160, width=50)
        self.b33 = ttk.Button(self.window7, text='Remove',  style='remove.TButton' , command=self.delete)
        self.b33.place(x=160, y=190, width=75)
        self.can = Canvas(self.window7, width=566.5, height=50, bg='#20232A')        
        self.can.place(x=5, y=225)
        self.bn = ttk.Button(self.window7, text='Load',  style='green/black.TButton' , command=self.letstrain)
        self.bn.place(x=340, y=240, width=100)
        self.can = Canvas(self.window7, width=176.5, height=50, bg='#20232A')        
        self.can.place(x=393.5, y=5)
        self.l56=Label(self.window7,text="Added Layers",font=('times new roman', 14, 'bold'),style='green/black.TLabel',background='#20232A',foreground="#dbdbdb")
        self.l56.place(x=410,y=20) 
        self.scrollbar1 = Scrollbar(self.window7)
        self.scrollbar1.pack( side = RIGHT, fill=Y )
        self.T2 =Listbox(self.window7,yscrollcommand = self.scrollbar1.set)
        self.scrollbar1.config(command = self.T2.yview)
        self.T2.place(x=395,y=60,width=180)
    
        
     
    def letstrain(self):
        self.window7.destroy()
        self.testtrainmodel(self.data,self.layer)
        
    def testtrainmodel(self,datapath,layer):
        model_new,c_c,activate,threshold=self.train(datapath,layer)
        self.model=model_new
        weights=self.saveweight(c_c,datapath,layer,model_new,activate,threshold)
        self.weights=weights
        self.test(model_new,datapath,weights)
        MsgBox = messagebox.askquestion ('Select an option','Do you want to test your model',icon = 'question')
        if MsgBox=="yes":
            self.testing=1
            self.choosetestfiles(model_new,weights)
        else:
            self.savedone()
   
    def saveweight(self,c_c,datapath,layer,model_new,activate,threshold):
        MsgBox = messagebox.askquestion ('Select an option','Do you want to save the weights',icon = 'question')
        if MsgBox=="yes":
            fileweightsave=filedialog.asksaveasfilename(initialdir = "/",title = "Select file",defaultextension=".hdf5")
            print(fileweightsave)
            loc=fileweightsave.split(".hdf5")
            file = open(loc[0]+'.txt','w') 
            file.write(str(c_c)+","+str(activate)+","+str(threshold)+"\n")
            i=1
            for i in range(0,len(datapath)):
                sdata=datapath[i].split('/')
                file.write(str(i+1)+","+str(sdata[len(sdata)-1])+"\n")
                i=i+1
            for i in range(0,len(layer)):
                file.write(layer[i]+"\n")        
            model_new.save_weights(fileweightsave,overwrite=True)
            return fileweightsave
        
    def addConv(self):
                self.layer.append("Conv")
                self.T2.config(state="normal")
                self.T2.delete('0', END)
                for ent in self.layer:
                    self.T2.insert(END, ent+'\n')
                self.T2.config(state="disabled")
                self.layercount=self.layercount+1
                
    def delete(self):
                
                self.layer=self.layer[:-1]
                self.T2.config(state="normal")
                self.T2.delete('0', END)
                for ent in self.layer:
                    self.T2.insert(END, ent+'\n')
                self.T2.config(state="disabled")
                self.layercount=self.layercount-1
            
    def addPool(self):
                if(self.value is 1):
                    self.layer.append("MaxPooling")
                else:
                    self.layer.append("AveragePooling")
                self.T2.config(state="normal")
                self.T2.delete('0', END)
                for ent in self.layer:
                    self.T2.insert(END, ent+'\n')
                self.T2.config(state="disabled")
                self.layercount=self.layercount+1
                
    def addLSTM(self):
                self.layer.append("LSTM")
                self.T2.config(state="normal")
                self.T2.delete('0', END)
                for ent in self.layer:
                    self.T2.insert(END, ent+'\n')
                self.T2.config(state="disabled")
                self.layercount=self.layercount+1
                
    def write(self):
        w=self.e1.get()
        print(w)
        r=self.e2.get()
        print(r)
        t=self.e3.get()
        act=self.variable.get()
        pool=str(self.variable1.get())
        poolc="MaxPooling"
        if(pool==poolc):
            self.value=1
        else:
            self.value=2
        print(self.value)
        value=str(self.value)
        self.file = open('load.txt','w')
        end='//null//'
        self.file.write('epoch:'+w+end)
        self.file.write('cls:'+r+end)
        self.file.write('threshold:'+t+end)
        self.file.write('activation:'+act+end)
        self.file.write('pooling:'+pool+end)
        self.file.write('poolingvalue:'+value+end)
        i=self.var3.get()
        self.file.write('imgpath:'+i+end)
        h=self.var4.get()
        self.file.write('headerpath:'+h+end) 
        self.row,self.col,self.band,self.dtypevalue=self.headerreader(self.head)
        self.file.write('row:'+str(self.row)+end)
        self.file.write('col:'+str(self.col)+end)
        self.file.write('band:'+str(self.band)+end)
        self.file.write('dtype:'+str(self.dtypevalue)+end)
        self.file.close()
        self.flag=1
        
    def read(self):
        w=self.e1.get()
        print(w)
        r=self.e2.get()
        print(r)
        t=self.e3.get()
        act=self.variable.get()
        pool=str(self.variable1.get())
        poolc="MaxPooling"
        if(pool==poolc):
            self.value=1
        else:
            self.value=2
        print(self.value)
        value=str(self.value)
        self.row,self.col,self.band,self.dtypevalue=self.headerreader(self.head)
        if (self.dtypevalue == 8):
            dtype=np.uint8
        elif(self.dtypevalue ==16):
            dtype=np.uint16
        return int(w),int(r)+1,float(t),act,pool,self.image,self.row,self.col,self.band,self.value,dtype,self.dtypevalue
                   
                
    
        ##########################
    def nextw(self):
        if len(self.e1.get()) == 0:
            messagebox.showerror("Error", "The value of epochs block is empty")
        elif len(self.e2.get()) == 0:
            messagebox.showerror("Error", "The Number of Training Classes block is empty")
        elif len(self.e3.get()) == 0:
            messagebox.showerror("Error", "The Threshold value block is empty")
        elif len(self.var4.get()) == 0:
            messagebox.showerror("Error", "Insert the Image file")
        elif len(self.var3.get()) == 0:
            messagebox.showerror("Error", "Insert the header file")
        else:
            self.write()
            if self.flag==0:
                messagebox.showerror("Error", "not written")
            self.window9=ttk.tkinter.Toplevel()
            self.window9.geometry("800x295")        
            self.can = Canvas(self.window9, width=785, height=50, bg='#20232A')        
            self.can.place(x=5, y=5)
            self.l11=Label(self.window9,text="Add the Training data",font=('times new roman', 14, 'bold'),style='green/black.TLabel',background='#20232A',foreground="#dbdbdb")
            self.l11.place(x=50,y=20)        
            self.can = Canvas(self.window9, width=785, height=180, bg='#20232A')        
            self.can.place(x=5, y=55)
            self.l5 = Label(self.window9,text="Load the Training data :",font=('times new roman', 12, 'bold'),style='gb.TLabel',background='#20232A',foreground="#dbdbdb")
            self.l5.place(x=35, y=65, width=350)
            self.b3 = Button(self.window9, text='Add', style='add.TButton',command=self.add )
            self.b3.place(x=45, y=105, width=100)
            self.b4 = Button(self.window9, text='Remove', style='remove.TButton',command=self.remove )
            self.b4.place(x=45, y=135, width=100)
            self.l5 = ttk.Label(self.window9,text="Training data :",font=('times new roman', 12, 'bold'),style='gb.TLabel',background='#20232A',foreground="#dbdbdb")
            self.l5.place(x=450, y=65, width=200)
            self.T =Text(self.window9, height=7, width=53)
            self.T.pack()
            self.T.place(x=352,y=100)
            self.T.config({"background": "#20232A"})
            self.T.config({"foreground":"#dbdbdb"})
            self.T.config({"border": 0})
            self.can = Canvas(self.window9, width=785, height=50, bg='#20232A')        
            self.can.place(x=5, y=235) 
            self.b7 = ttk.Button(self.window9, text='Next', style='green/black.TButton',command=self.addlayers)
            self.b7.place(x=625, y=245, width=110) 
            
    def ReadBilFile(self,bil,bands,pixels,dtype):
        extract_band = 1
        image = np.zeros([pixels, bands], dtype=np.uint8)
        gdal.GetDriverByName('EHdr').Register()
        img = gdal.Open(bil)
        while bands >= extract_band:
            bandx = img.GetRasterBand(extract_band)
            datax = bandx.ReadAsArray()
            temp = datax
            store = temp.reshape(pixels)
            for i in range(pixels):
                image[i][extract_band - 1] = store[i]
            extract_band = extract_band + 1
        return image

    
        
    def train(self,datapath,layer):  
        epoch,c_c,threshold,activate,pool,imgpath,row,col,bands,value,dtype,d = self.read()
        values = []
        path=datapath
        c_l={}
        clicks={}
        c=0
        layercount=len(layer)
       
    ##addition
        for add in path:
            c=int(c)+1
            print("{} class {}".format(add,c))  #E:/Data/sub1 class 1
            c_l[add]=c
    ###
        for address in path:
            with open(address, "rb") as f:
                k = len(f.read())
                clicks[address] = (k // 2 // bands) if (k // 2 // bands) < 400 else (k // 2 // bands) // 4
                print('{} ==> {}'.format(address, clicks[address]))
             
        for address in path:
            with open(address, "rb") as f:
                b = array.array("H")
                b.fromfile(f, clicks[address]*bands)
                if sys.byteorder == "little":
                    b.byteswap()
                for v in b:
                    values.append(v)


        ll = (len(values))
        rex = ll // bands
        print(ll, rex)
        
        f_in = np.zeros([ll], dtype)
        x = 0
        for i in range(ll):
            f_in[x] = values[i]
            x += 1

   
        y_train = np.zeros([rex], dtype)
    
        mark = 0
        for add in path:
            for i in range(clicks[add]):
                y_train[mark+i] = c_l[add]
            mark = mark + clicks[add]

        
        x_train = f_in.reshape(rex, bands)

        seed = 6
        np.random.seed(seed)

        x_train = x_train / 2**(d-1)
        
        num_pixels = bands

        for v in y_train:
            print(v, end=" ")

        y_train = np_utils.to_categorical(y_train)
        n_classes = c_c
        print(x_train)
        print(20*'#')
        print(y_train)

        print(x_train.shape)
        print(y_train.shape)


        X = x_train.reshape(x_train.shape[0], bands, 1)
        n_units=128
        n_classes=c_c
        batch_size=50
        j=3
        t=int(len(layer))-1
    ########
       

        model = Sequential()
        for i in range(0,layercount):
            if(layer[i]=="Conv"):
                if(i==0):
                    model.add(Conv1D(2 ** j, 2, activation=activate, padding='same', input_shape=[bands, 1]))
                else:
                    model.add(Conv1D(2 ** j, 2, activation=activate, padding='same'))
            
                j=j+1
            elif(layer[i]=="MaxPooling" and i!=0):
                model.add(MaxPooling1D(2))
            elif(layer[i]=="AveragePooling" and i!=0):
                model.add(AveragePooling1D(2))
            elif(layer[i]=="LSTM"):
                if(i==0):
                    model.add(LSTM(2**j, return_sequences=False,input_shape=[bands, 1]))
                elif(i==t):
                    model.add(LSTM(2**(j-1)))
                else:
                    model.add(LSTM(2**j, return_sequences=True))
                    j=j+1
    
        model.add(Dense(n_classes, activation='sigmoid'))
        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
        model.summary()
        model.fit(X, y_train, batch_size=50, epochs=epoch)  
        return model,c_c,activate,threshold
   
                              
    ################to only test                          
    def test(self,model_new,datapath,weights): 
        train_cycle,c_c,threshold,activate,pool,imgpath,row,col,bands,value,dtype,d=self.read()  
        pixels = row * col
        y_test = np.zeros([row * col], dtype)
        x_test = self.ReadBilFile(imgpath, bands, pixels,dtype)
        x_test = x_test.reshape(row*col, bands, 1)
    
        x_test = x_test / (2**(d-1))
        y_test = np_utils.to_categorical(y_test)
    
        datapathname=[]
        for i in range(0,len(datapath)):
            sdata=datapath[i].split('/')
            print(sdata)
            datapathname.append(sdata[len(sdata)-1])
        model_new.load_weights(weights)
        y_test_new = model_new.predict(x_test)
        maxlim=threshold
        for rno in range(1,pixels):
            for cno in range(1,c_c-1):
                if(y_test_new[rno][cno]<maxlim):
                    y_test_new[rno][cno]=0;

        print( 'printing new values of ytestnew' )
        print(y_test_new)
        print(" new values printed")
        print(20*'%')
        print(y_test_new.shape)
        print(20*'%')
        y_test1 = np.argmax(y_test_new, axis=1)
        print(30*'*')
        print("this is predicted output")
        imageresult=[]
        imgname=[]
    
        k=y_test1.reshape(row,col)
        plt.imshow(k)
        plt.show()
        result = Image.fromarray((k * (2**(d-1))//c_c).astype(dtype))
        imageresult.append(result)
        name="hard"
        imgname.append(name)
        self.putimage(result)
        
        if self.testing==1:
            MsgBox = messagebox.askquestion ('Select an option','Do you want to save the classified',icon = 'question')
            if MsgBox=="yes":
                loc=filedialog.askdirectory()
                result.save(loc+'/'+'hardtested'+'_classified.tiff')
        
        
        if self.testing==0:
            for i in range(1, c_c):
                img = y_test_new[:,i].reshape(row, col)
                plt.imshow(img*(2**(d-1)))
                plt.colorbar()
                plt.show()
                result = Image.fromarray(((img * (2**(d-1)))).astype(dtype))
                imageresult.append(result)
                imgname.append(datapathname[i-1])
                self.putimage(result,i)        
                
            MsgBox = messagebox.askquestion ('Select an option','Do you want to save the images',icon = 'question')
            if MsgBox=="yes":
                loc=filedialog.askdirectory()
                for i in range (0,len(imageresult)):
                    imageresult[i].save(loc+'/'+imgname[i]+'_classified.tiff')                      
    
                              
                              
                              
                              
                              
                              
                              
                              ############
    
                          
    def putimage(self,result,window3=10):     
        self.window3=tkinter.Toplevel()
        self.frame = Frame(self.window3, relief=SUNKEN)
        self.frame.grid_rowconfigure(0, weight=1)
        self.frame.grid_columnconfigure(0, weight=1)                       
        img = ImageTk.PhotoImage(result)
               
        self.label = Label(self.frame,image=img)
        self.label.grid(row=0, column=0, sticky=N+S+E+W)
  
        self.label.image=img

     
        self.frame.pack(fill=BOTH,expand=1) 
        
    def dotesting(self):
        self.test(self.model,self.data,self.weights)
        
    def choosetestfiles(self,model_new,weights):
        
        self.window7.destroy()
        self.window=ttk.tkinter.Toplevel()
        self.window.geometry("500x300")
        self.can = Canvas(self.window, width=500, height=300, bg='#20232A')        
        self.can.place(x=2.5, y=2.5) 
        self.bf1 = ttk.Button(self.window, text='Select ImageFile', style='green/black.TButton' , command=self.img )
        self.bf1.place(x=15, y=35, width=110)
        ttk.Label(self.window,textvariable=self.var4,relief="sunken").place(x=145, y=35, width=300)
        self.bf2 = ttk.Button(self.window, text='Select HeaderFile', style='green/black.TButton' , command=self.header)
        self.bf2.place(x=15, y=100, width=110)
        ttk.Label(self.window,textvariable=self.var3,relief="sunken" ).place(x=145, y=100, width=300)      
        self.bf3 = ttk.Button(self.window, text='Test', style='green/black.TButton' , command=self.dotesting)
        self.bf3.place(x=100, y=200, width=110)
        self.bf4 = ttk.Button(self.window, text='Exit', style='green/black.TButton' , command=self.savedone)
        self.bf4.place(x=270, y=200, width=110)
        
    
                                
    
        
     
            
            
if __name__ == '__main__':
        root = Tk()
        root.resizable(0, 0)
        obj = MyGUI(root)
        root.mainloop()
```
